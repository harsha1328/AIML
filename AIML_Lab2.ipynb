```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')
plt.xlabel("Years")
plt.ylabel("Yields")
plt.title("Crop yields in kanto")
plt.legend(["Apples","Oranges"]);
```

```python
year=range(2000,2008)
apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
plt.plot(year,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy')
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5)
plt.xlabel("Years")
plt.ylabel("Yields")
plt.title("Crop yields in kanto")
plt.legend(["Apples","Oranges"]);
```

```python
plt.plot(year,apples,'s-b')
plt.plot(year,oranges,'o--r')
plt.xlabel("Years")
plt.ylabel("Yields")
plt.title("Crop yields in kanto")
plt.legend(["Apples","Oranges"]);
```

```python
plt.plot(year,apples,'or')
plt.title('yield of apples')
```

```python
plt.figure(figsize=(10,6))
plt.plot(year,apples,'or')
plt.title('yield of apples')
```

```python
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title('Bar chart')
plt.legend(['Bar'])
plt.show()
```

```python
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.scatter(x,y)
plt.legend("A")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot')
plt.show()
```

```python
x=[1,2,3,4]
e=(0.1,0,0,0)
plt.pie(x,explode=e)
plt.title("Pie chart")
plt.show()
```
